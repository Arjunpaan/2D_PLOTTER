```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def send_command(command):
    print(command)  # Just print the command instead

def move_pen(x, y):
    send_command(f"X{x}")
    send_command(f"Y{y}")

def pen_up():
    send_command("U")

def pen_down():
    send_command("D")

# Load the image
image = cv2.imread('/content/ex image.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image loaded correctly
if image is None:
    print("Error: Image not found or failed to load.")
else:
    # Threshold the image to binary
    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)

    # Ensure binary_image is of type uint8
    binary_image = binary_image.astype(np.uint8)

    # Find contours
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Loop through each contour and send commands to the plotter
    for contour in contours:
        for point in contour:
            x, y = point[0]
            move_pen(x, y)
            pen_down()
        pen_up()

    # Visualize the result (Optional for testing in Colab)
    plt.imshow(binary_image, cmap='gray')
    plt.title('Binary Image with Contours')
    plt.show()
```
